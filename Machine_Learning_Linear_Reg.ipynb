```python
df_num=df.select_dtypes(exclude='object')
df_cat=df.select_dtypes(include='object')
df_cat
```

```python
df_cat['Eduacation'].unique()
```

```python
df_cat['Eduacation']=df_cat['Eduacation'].replace({'LessThanHighSchool':0,'Intermediate':1,'HighSchool':2,'graduate':3,'PostGraduate':4})
df_cat.head(3)
```

```python
df.dtypes
```

```python
df_dummies = pd.get_dummies(df_cat,drop_first=True)
```

```python
df_dummies.head(3)
```

```python
df_new=pd.concat([df_num,df_dummies],axis=1)
df_new.sample(5)
```

```python
x= df_new.drop(['Earnings_1978'],axis=1)
type(x)
```

```python
y=df_new['Earnings_1978']
y
type(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8, random_state= 8)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
# y = mx + c
lr.fit(x_train,y_train)
```

```python
lr.coef_ # m
```

```python
lr.intercept_ # c
```

```python
df_new.head(2)
```

```python
# (-1.02560058e+02*45)+ (6.72375386e+02*1) + (2.71637643e-01*21516.670) + (4.32381877e-01*25243.550) + 7688.2392691370505
```

```python
y_pred= lr.predict(x_test)
y_pred.shape
```

```python
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt
rmse=sqrt(mean_squared_error(y_test,y_pred))
r_squared=r2_score(y_test,y_pred)
print(rmse)
print(r_squared)
```

```python
residual = y_test - y_pred
sns.histplot(residual)
```

```python
# StandardScaler = X-mean/std
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test) # we dont use fit because fit means std and mean value ki value nhi pata hoti in test , jo bhi naya data aata rahega to usse change krte rahunga
```

```python
lr.fit(X_train,y_train)
Y_pred1 = lr.predict(X_test)
rmse=sqrt(mean_squared_error(y_test,Y_pred1))
r_squared=r2_score(y_test,Y_pred1)
print(rmse)
print(r_squared)
```
